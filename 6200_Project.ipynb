#Setup Nova-Pro query and Bedrock client

```python
import os
import json
import re

client = boto3.client("bedrock-runtime", region_name="us-virginia-1")

max_tokens = 512
model_id = "meta.llama3-1-405b-instruct-v1:0"
model_endpoint = ""

def query_nova_pro(question, context=None):
    prompt = f"Answer the following question based on the context provided below.\n\nContext: {context}\n\nQuestion: {question}\nAnswer:"

    response = client.invoke_model(
        ModelId=model_endpoint,  # Replace this with your actual Nova Pro model endpoint
        Body=json.dumps({"prompt": prompt}),
        ContentType="application/json"
    )

    result = json.loads(response["Body"].read())
    answer = result.get("generated_text", "").strip()
    return answer
```

#Eval Script for Natural-Questions

```python
# Evaluation functions adapted from nq_eval.py: https://github.com/google-research-datasets/natural-questions/blob/master/nq_eval.py

def normalize_answer(s):
    def remove_articles(text):
        return re.sub(r'\b(a|an|the)\b', ' ', text)
    def white_space_fix(text):
        return ' '.join(text.split())
    def remove_punc(text):
        return re.sub(r'[^\w\s]', '', text)
    def lower(text):
        return text.lower()
    return white_space_fix(remove_articles(remove_punc(lower(s))))

def compute_exact(a_gold, a_pred):
    return int(normalize_answer(a_gold) == normalize_answer(a_pred))

def compute_f1(a_gold, a_pred):
    gold_tokens = normalize_answer(a_gold).split()
    pred_tokens = normalize_answer(a_pred).split()

    if len(gold_tokens) == 0 or len(pred_tokens) == 0:
        # If either is empty:
        return int(gold_tokens == pred_tokens)

    common = set(gold_tokens) & set(pred_tokens)
    if len(common) == 0:
        return 0.0

    precision = len(common) / len(pred_tokens)
    recall = len(common) / len(gold_tokens)
    return 2 * precision * recall / (precision + recall)
```

#Compute Baseline:

```python
# Baseline Evaluation Loop
from tqdm import tqdm

total_em = 0
total_f1 = 0
processed_examples = 0

for example in tqdm(nq_dataset):
    question = example["question"]
    context = example.get("document_text", "")

    gold_answers = []
    if "annotations" in example:
        for annotation in example["annotations"]:
            if "short_answers" in annotation and len(annotation["short_answers"]) > 0:
                for ans in annotation["short_answers"]:
                    gold_answers.append(ans["text"])

    if not gold_answers:
        continue

    # Query Nova Pro (you might decide to include or exclude context)
    pred_answer = query_nova_pro(question, context)

    # Compare exact match and f1 scores
    em_scores = [compute_exact(gold, pred_answer) for gold in gold_answers]
    f1_scores = [compute_f1(gold, pred_answer) for gold in gold_answers]

    best_em = max(em_scores)
    best_f1 = max(f1_scores)

    total_em += best_em
    total_f1 += best_f1
    processed_examples += 1

overall_em = (total_em / processed_examples) * 100
overall_f1 = (total_f1 / processed_examples) * 100

print(f"Nova Pro Baseline on Natural Questions:\nExact Match: {overall_em:.2f}%\nF1 Score: {overall_f1:.2f}%")
```
